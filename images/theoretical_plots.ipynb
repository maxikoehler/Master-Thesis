# Theoretical Plots

## Imports and Definitions

```python
%matplotlib inline
import ipympl

import os
import sys
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl

ees_blue = [0 / 255, 112 / 255, 192 / 255]
ees_yellow = [213 / 255, 223 / 255, 0 / 255]
ees_green = [0 / 255, 166 / 255, 74 / 255]
ees_red = [192 / 255, 0 / 255, 0 / 255]
ees_lightblue = [0 / 255, 176 / 255, 240 / 255]
ees_black = [0 / 255, 0 / 255, 0 / 255]


# Set the matplotlib settings for plotting with LaTeX
plt.rcParams.update({
    "font.family": "Charter",
    "font.size": 12,
    "figure.autolayout": True,
    'text.usetex': True,
})
mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=[ees_blue, ees_yellow, ees_green, ees_red, ees_lightblue])
save_path = os.path.join(os.getcwd(), 'plots/')

if not os.path.isdir(save_path):
    os.makedirs(save_path)

print(save_path)
```

## Trigonometric Functions

```python
angle = np.linspace(-np.pi, np.pi, 1000)
angle_tan = np.linspace(-np.pi/2, np.pi/2, 1000)
angle_tan2 = np.linspace(-np.pi/2-0.001, -np.pi, 1000)
angle_tan3 = np.linspace(np.pi/2+0.001, np.pi, 1000)
angle_deg = angle * 180/np.pi

sin = np.sin(angle)
cos = np.cos(angle)
tan = np.tan(angle_tan)
tan2 = np.tan(angle_tan2)
tan3 = np.tan(angle_tan3)

def deg2rad(x):
    return x * np.pi / 180


def rad2deg(x):
    return x * 180 / np.pi

fig, ax = plt.subplots(layout='constrained', figsize=(8,6))

ax.plot(angle, sin, label=r'$\sin$')
ax.plot(angle, cos, label=r'$\cos$')
ax.plot(angle_tan, tan, color=ees_green, label=r'$\tan$')
ax.plot(angle_tan2, tan2, color=ees_green)
ax.plot(angle_tan3, tan3, color=ees_green)

ax.set_xlabel('Angle in Radians')
ax.set_ylabel('Function Value')
ax.set_ylim(-3,3)
ax.grid()
ax.legend()

secax = ax.secondary_xaxis('top', functions=(rad2deg, deg2rad))
secax.set_ticks([-180, -150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180])
secax.set_xlabel('Angle in Degrees')

plt.savefig(save_path + 'trigonometric_functions.pdf')
plt.show()
```
